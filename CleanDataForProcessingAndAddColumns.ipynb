```python
#Cargamos datos - Mode
import datetime
import pandas as pd

#Creación de una Lista
meses = []
for mes in range(1,13):
    meses.append(pd.read_csv("2019_{}.csv".format(mes)))
    
#Concatenación de los Elementos de la Lista a un DataFrame
datos_mode_df = pd.concat(meses)
datos_mode_df.head()
```

```python
#Conversión de las Columnas con Fechas al formato de serie de tiempo - Mean
datos_mean_df["Fecha Inicio Viaje"] = pd.to_datetime(datos_mean_df["Fecha Inicio Viaje"],dayfirst=True)
datos_mean_df["Fecha Fin Viaje"] = pd.to_datetime(datos_mean_df["Fecha Fin Viaje"],dayfirst=True)
#Agregar columna de viajes
datos_mean_df["Viaje"] = 1
datos_mean_df.head()
```

```python
#Conversión de las Columnas con Fechas al formato de serie de tiempo - Mode
datos_mode_df["Fecha Inicio Viaje"] = pd.to_datetime(datos_mode_df["Fecha Inicio Viaje"],dayfirst=True)
datos_mode_df["Fecha Fin Viaje"] = pd.to_datetime(datos_mode_df["Fecha Fin Viaje"],dayfirst=True)
#Agregar columna de viajes
datos_mode_df["Viaje"] = 1
datos_mode_df.head()
```

```python
#Sacamos la diferencia del tiempo - Mean
# Determinar el tiempo de recorrido de cada viaje realizado por cada usuario

#Apagar los Warnings
pd.set_option("mode.chained_assignment", None)

datos_df_tiempo = datos_mean_df[["Hora Inicio Viaje", "Hora Fin Viaje"]]

datos_df_tiempo["Hora Inicio Viaje"] = pd.to_datetime(datos_df_tiempo["Hora Inicio Viaje"])
datos_df_tiempo["Hora Fin Viaje"] = pd.to_datetime(datos_df_tiempo["Hora Fin Viaje"])

datos_mean_df['Tiempo'] = datos_df_tiempo["Hora Fin Viaje"] - datos_df_tiempo["Hora Inicio Viaje"]
datos_mean_df
```

```python
#drop NA
datos_mean_df.dropna(inplace = True)
datos_mode_df.dropna(inplace = True)
```

```python
#Sacamos la diferencia del tiempo - Mode
# Determinar el tiempo de recorrido de cada viaje realizado por cada usuario

#Apagar los Warnings
pd.set_option("mode.chained_assignment", None)

datos_df_tiempo = datos_mode_df[["Hora Inicio Viaje", "Hora Fin Viaje"]]

datos_df_tiempo["Hora Inicio Viaje"] = pd.to_datetime(datos_df_tiempo["Hora Inicio Viaje"])
datos_df_tiempo["Hora Fin Viaje"] = pd.to_datetime(datos_df_tiempo["Hora Fin Viaje"])

datos_mode_df['Tiempo'] = datos_df_tiempo["Hora Fin Viaje"] - datos_df_tiempo["Hora Inicio Viaje"]
datos_mode_df
```

```python
#Convertir tiempo en float - Mean
datos_mean_df['Tiempo'] =datos_mean_df['Tiempo'].astype('timedelta64[m]')
```

```python
#Convertir tiempo en float - Mode
datos_mode_df['Tiempo'] =datos_mode_df['Tiempo'].astype('timedelta64[m]')
```

```python
#Agrupa - Mean
records_df_mean = datos_mean_df.groupby(datos_mean_df["Fecha Inicio Viaje"]).agg({'Fecha Inicio Viaje':'first',
                                                                                  'Fecha Fin Viaje':'first',
                                                                                  'ID Origen':pd.Series.mode,
                                                                                  'ID Destino':pd.Series.mode,
                                                                                  'Género': pd.Series.mode,
                                                                                  'Tiempo':'mean',
                                                                                  'Viaje':'sum'}) 
```

```python
#Agrupar - Mode
records_df_mode = datos_mode_df.groupby(datos_mode_df["Fecha Inicio Viaje"]).agg({'Fecha Inicio Viaje':'first',
                                                                   'Fecha Fin Viaje':'first',
                                                                   'ID Origen':pd.Series.mode,
                                                                   'ID Destino':pd.Series.mode,
                                                                    'Género': pd.Series.mode,
                                                                   'Tiempo':pd.Series.mode,
                                                                   'Viaje':'sum'}) 
```

```python
#Calular and add season - Mean
#Add season column
'''
1 - Winter
2 - Spring
3 - Summer
4 - Fall
'''
season =[]
for i in records_df_mean["Fecha Inicio Viaje"]:
    month = i.month
    if (month > 11 or month <= 3):
        season.append(1)
    elif (month == 4 or month == 5):
        season.append(2)
    elif (month>=6 and month <=9):
        season.append(3)
    else:
        season.append(4)
#Insert column
records_df_mean.insert(2,'Temporada',season,True)
```

```python
#Calcular and add season - Mode
#Add season column
'''
1 - Winter
2 - Spring
3 - Summer
4 - Fall
'''
season =[]
for i in records_df_mode["Fecha Inicio Viaje"]:
    month = i.month
    if (month > 11 or month <= 3):
        season.append(1)
    elif (month == 4 or month == 5):
        season.append(2)
    elif (month>=6 and month <=9):
        season.append(3)
    else:
        season.append(4)
#Insert column
records_df_mode.insert(2,'Temporada',season,True)
```
