```python
import datetime
import pandas as pd

#Creación de una Lista
meses = []
for mes in range(1,13):
    meses.append(pd.read_csv("2019_{}.csv".format(mes)))
    
#Concatenación de los Elementos de la Lista a un DataFrame
datos_df = pd.concat(meses)
datos_df.head()
```

```python
datos_df.head()
```

```python
#Conversión de las Columnas con Fechas al formato de serie de tiempo
datos_df["Fecha Inicio Viaje"] = pd.to_datetime(datos_df["Fecha Inicio Viaje"],dayfirst=True)
datos_df["Fecha Fin Viaje"] = pd.to_datetime(datos_df["Fecha Fin Viaje"],dayfirst=True)
```

```python
datos_df["Viaje"] = 1
```

```python
datos_df.head()
```

```python
datos_df.dropna(inplace = True)
datos_df.isnull().sum()
```

```python
datos_df.shape
```

```python
month = pd.DatetimeIndex(datos_df["Fecha Inicio Viaje"]).month
```

```python
print(len(month))
```

```python
#New variables for every month trip count
tJan = 0
tFeb = 0
tMar = 0
tApr = 0
tMay = 0
tJun = 0
tJul = 0
tAug = 0
tSep = 0
tOct = 0
tNov = 0
tDec = 0

for i in datos_df["Fecha Inicio Viaje"]:
    month = i.month
    if month == 1:
        tJan = tJan+1
    elif month == 2:
        tFeb = tFeb+1
    elif month == 3:
        tMar = tMar+1
    elif month == 4:
        tApr = tApr+1
    elif month == 5:
        tMay = tMay+1
    elif month == 6:
        tJun = tJun+1
    elif month == 7:
        tJul = tJul+1
    elif month == 8:
        tAug = tAug+1
    elif month == 9:
        tSep = tSep+1
    elif month == 10:
        tOct = tOct+1
    elif month == 11:
        tNov = tNov+1
    elif month == 12:
        tDec = tDec+1        
```

```python
data = [['Enero',tJan],['Febrero',tFeb],['Marzo',tMar],['Abril',tApr],['Mayo',tMay],['Junio',tJun],['Julio',tJul],['Agosto',tAug],
        ['Septiembre',tSep],['Octubre',tOct],['Noviembre',tNov],['Diciembre',tDec]]
#Append values to new DF
records_df = pd.DataFrame(data,columns=['Mes','Viajes'])
records_df
```

```python
#https://stackoverflow.com/questions/46444444/how-to-calculate-sum-of-column-per-day-in-pandas
records_df = datos_df.groupby(datos_df["Fecha Inicio Viaje"]).agg({'Fecha Inicio Viaje':'first','Viaje':'sum'}) #
```
